```python
# an identity matrix of size m
Identity = np.ones(m)

# all the cp variables
X = cp.Variable((m, 1)) # m investment types
Bank = cp.Variable((n, 1)) # bank account balance
Withdraw = cp.Variable((n, 1)) # withdrawal at each time 't'

# the constraints for the problem
constraints = [X >= 0, Bank >= 0, Withdraw >= 0, (Withdraw - e + cp.matmul(P, X)) >= 0]

# for each iteration ,i.e., time t
for i in range(n-1):
  constraints = constraints +  [Bank[i+1] == (1 + rho)*Bank[i] - Withdraw[i]]

# objective of the problem
objective = cp.Minimize(Bank[0] + cp.matmul(Identity.T, X))
problem = cp.Problem(objective, constraints)
expense_stream = problem.solve()
print(expense_stream)
```
