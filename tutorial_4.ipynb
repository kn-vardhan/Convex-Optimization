```python
# Convex Optimization Tutorial 4
# Polynomial Approximation
# The Second Cell is the sum of squares of the differences. First Cost Function.
# The Third Cell is for plotting the graph between Cost Function and n for values of n from 1 to 30.
# The Fourth Cell is the sum of the absolute values of the differences. Second Cost Function
# The Fifth Cell is for plotting the graph between Cost Function and n for values of n from 1 to 30.

import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
n = [5, 10, 20] # for all the 3 test cases
N = [25, 100, 400] # corresponding 'N' value of the 3 test cases

for t in range(3):
  n1 = n[t]
  N1 = N[t]
  M = 2*N1 + 1
  x_k = np.linspace(-1, 1, M)
  x_k_i = []
  for i in range(n1):
    x_k_i.append(pow(x_k, i))

  x_k_i = np.array(x_k_i)
  A = cp.Variable([n1, 1])

  fx = np.power(x_k, n1).reshape(1, M)

  objective = cp.sum_squares(fx - cp.matmul(A.T, x_k_i))

  problem = cp.Problem(cp.Minimize(objective))
  cost_fn = problem.solve()
  print(f'TEST CASE {t+1} ==> n = {n1}, N = {N1}')
  print(f'The minimized cost function value is {cost_fn}.')
  print(f"The coefficients of the n-1 degree polynomial for n = {n1} are")
  print(A.value, end='\n\n')
  
  plt.figure()
  plt.plot(x_k, fx.T, label='true')
  plt.plot(x_k, (np.matmul(A.value.T, x_k_i)).T, label='approx')
  plt.legend()
```
